```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# разбейте исходную таблицу на 2: по полу пассажиров
# удалите пустые строки или заполните их соответствующими символами
# Построить гистограмму распределения пассажиров по возрасту
# Вычислите мин и макс, среднее, моду, медиану возраста пассажира каждого пола
# визуализировать возраст при помощи различных настроек метода plot

df = pd.read_csv('titanic.csv').dropna().drop_duplicates()
```

### Мужчины

```python

male_df = df[df['Sex'] == 'male']

male_df
```

```python
male_df['Age'].hist()
```

```python
print(male_df['Age'].describe())
print(f'mode: {male_df['Age'].mode()[0]}')
print(f'median: {male_df['Age'].median()}')
```

### Женщины

```python
female_df = df[df['Sex'] == 'female']

female_df
```

```python
female_df['Age'].hist()
```

```python
print(female_df['Age'].describe())
print(f'mode: {female_df['Age'].mode()[0]}')
print(f'median: {female_df['Age'].median()}')
```

### Графики

```python
male_df['Age'].plot(kind='hist', bins=20, color='blue', label='Male')
female_df['Age'].plot(kind='hist', alpha=0.7, bins=20, color='red', label='Female')

plt.title('Hist')
plt.xlabel('Age')
plt.legend()
plt.show()
```

```python
plt.scatter(male_df.index, male_df['Age'], alpha=0.7, color='blue', label='Male')
plt.scatter(female_df.index, female_df['Age'], alpha=0.7, color='red', label='Female')

plt.title('Scatter')
plt.ylabel('Age')    
plt.legend()
plt.show()
```

```python
male_df['Age'].plot(kind='density', color='blue', label='Male')
female_df['Age'].plot(kind='density', color='red', label='Female')

plt.title('Density')
plt.xlabel('Age')
plt.legend()
plt.show()
```
